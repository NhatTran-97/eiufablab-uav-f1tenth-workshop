```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
angle = 45 * np.pi/180

sin = np.sin(angle)
cos = np.cos(angle)
tan = np.tan(angle)
print(f"sin({angle}) = {sin}")
print(f"cos({angle}) = {cos}")
print(f"tan({angle}) = {tan}")

arc_sin = np.arcsin(sin) * 180/np.pi
arc_cos = np.arcsin(cos) * 180/np.pi
arc_tan = np.arcsin(tan) * 180/np.pi
print()
print(f"arcsin({sin}) = {arc_sin}")
print(f"arccos({cos}) = {arc_cos}")
print(f"arctan({tan}) = {arc_tan}")


```

```python
# use atan2
atan2 = np.arctan2(1, 1) * 180/np.pi
print(f"atan2(1, 1) = {atan2}")
atan2 = np.arctan2(1, -1) * 180/np.pi
print(f"atan2(-1, 1) = {atan2}")
atan2 = np.arctan2(-1, -1) * 180/np.pi
print(f"atan2(-1, -1) = {atan2}")
atan2 = np.arctan2(-1, 1) * 180/np.pi
print(f"atan2(1, -1) = {atan2}")
```
