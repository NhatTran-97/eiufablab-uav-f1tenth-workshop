```python
class MyRobot:
    def __init__(self, name):
        # ====== ATTRIBUTES ======
        self.name = name
        self.x = 0.0
        self.y = 0.0
        self.yaw = 0.0
        self.battery = 100
        self.state = "IDLE"

    # ====== ACTIONS ======
    def move_forward(self, distance):
        if self.battery <= 0:
            print("Battery empty")
            return
        self.x += distance
        self.battery -= distance * 0.5
        self.state = "NAVIGATING"

        print(f"{self.name} moved forward {distance} m")

    def rotate(self, angle):
        self.yaw += angle
        self.battery -= abs(angle) * 0.1
        print(f"{self.name} rotated {angle} degrees")

    def stop(self):
        self.state = "IDLE"
        print("Robot stopped")

    def status(self):
        print("===== ROBOT STATUS =====")
        print(f"Name     : {self.name}")
        print(f"Position : ({self.x:.2f}, {self.y:.2f})")
        print(f"Yaw      : {self.yaw:.2f}")
        print(f"Battery  : {self.battery:.1f}%")
        print(f"State    : {self.state}")
```
