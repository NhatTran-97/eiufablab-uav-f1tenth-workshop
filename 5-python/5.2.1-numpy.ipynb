```python
import numpy as np

angle = np.deg2rad(45)

print(f"np.sin(angle): {np.sin(angle)}")

print(f"np.cos(angle): {np.cos(angle)}")

print(f"np.tan(angle): {np.tan(angle)}")

cot_angle = 1 / np.tan(angle)
print(f"cot(angle): {cot_angle}")

```
