```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/advertising.csv")
TV_data = df["TV"].values

plt.plot(TV_data)
plt.title("TV Advertising Data")
plt.xlabel("Index")
plt.ylabel("TV")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("advertising.csv")
TV_data = df["TV"].values

plt.scatter(range(len(TV_data)), TV_data)
plt.title("TV Advertising Data")
plt.xlabel("Row Index")
plt.ylabel("TV Budget")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("advertising.csv")

plt.plot(df["TV"], label="TV")
plt.plot(df["Radio"], label="Radio")
plt.plot(df["Newspaper"], label="Newspaper")
plt.plot(df["Sales"], label="Sales")

plt.title("Advertising Dataset")
plt.xlabel("Row Index")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("advertising.csv")

corr = df.corr()

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("advertising.csv")

plt.hist(df["TV"], bins=10)
plt.title("Distribution of TV Values")
plt.xlabel("TV")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
