```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "data/advertising.csv"
```

```python
data = np.genfromtxt(path, delimiter = ',', skip_header = True) 
```

```python
X = data[:,:-1]
y = data[:,-1]
```

```python
n_feature = X.shape[1]
print(n_feature)
label_feature, label_predict = ["TV", "RADIO", "PAPER"], "Sale"
fig, axs = plt.subplots(nrows = n_feature, ncols = 1, figsize = (30,30))
for i in range(n_feature):
    axs[i].scatter(X[:, i], y, alpha = 0.70)
    axs[i].set_xlabel(label_feature[i], fontsize = 'large', fontweight='bold')
    axs[i].set_ylabel(label_predict, fontsize ='larger', fontweight='bold')
    
    axs[i].spines['top'].set_visible(False)
    axs[i].spines['right'].set_visible(False)
fig.tight_layout()    
```

```python
X.shape, y.shape
```

<p><strong>Sale = w1*x1 + w2x2 + w3x3 +b</strong></p>

$y = W*x => y*X^-1 = W*x*x^-1 => W = (x*x^-1)^T*y*x^-1 $

```python
X = np.concatenate((X, np.ones((len(X),1))), axis = 1)
```

```python
def find_theta(X,y):
    return np.dot(np.dot(np.linalg.inv(np.dot(X.T,X)),X.T),y)
```

```python
w = find_theta(X,y)
```

```python
w.shape, X.shape, y.shape
```

```python
y_predict = X.dot(w)
```

```python
np.abs(y_predict - y).mean()
```

```python
for i, j in zip(y_predict,y):
    print(f"{i:.2f} - {j}")
```
