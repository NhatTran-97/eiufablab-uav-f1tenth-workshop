## Drawing y=ax+b Linear function, with x in range (0, 10)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 10)
print("x: ", x)

# Linear function: y = 2*x + 1
y = 2 * x + 1
print("y: ", y)
plt.plot(x, y)
plt.title("Graph of y = 2x + 1")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(False)
plt.show()
```

## Drawing y=ax^2 + bx + c Quadratic function, with x in range (0, 10)

```python
import numpy as np
import matplotlib.pyplot as plt

# x range
x = np.linspace(-10, 10, 50)

# Quadratic function: y = ax^2 + bx + c
a, b, c = 2, 3, 1
y = a*x**2 + b*x + c

plt.plot(x, y)
plt.title(f"Graph of y = {a}x^2 + {b}x + {c}")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(False)
# plt.axhline(0)  # optional: x-axis
# plt.axvline(0)  # optional: y-axis
plt.show()
```

## Drawing cosine function, with x in range (0, 2Pi)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 20)
y = np.cos(x)

plt.plot(x, y)
plt.title("Cosine Function")
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.grid(True)
plt.show()
```
