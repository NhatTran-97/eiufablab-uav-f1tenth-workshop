```python
import math
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from matplotlib.patches import Rectangle
import matplotlib.transforms as transforms
from IPython.display import HTML

# =========================
# Initial settings
# =========================
initial_position = [0.0, 0.0]
goal_position = [2.0, 2.0]
robot_size = 0.3
arrow_length = 0.5

# Number of frames
rotate_frames = 60
move_frames = 100
total_frames = rotate_frames + move_frames

# =========================
# Geometry calculation
# =========================
dx = goal_position[0] - initial_position[0]
dy = goal_position[1] - initial_position[1]

theta = math.atan2(dy, dx)              # rad
theta_deg = math.degrees(theta)         # deg
#distance = math.hypot(dx, dy)           # distance between 2 points
distance = math.sqrt(dx**2 + dy**2)

print(f"Theta = {theta_deg:.2f} deg")
print(f"Distance = {distance:.3f}")

# =========================
# Figure setup
# =========================
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_xlim(-0.5, 6)
ax.set_ylim(-0.5, 6)
ax.set_aspect('equal')
ax.grid(True)
ax.set_title("Robot Navigation Simulation")

# Draw start and goal
start_point, = ax.plot(initial_position[0], initial_position[1], 'bo', label='Start')
goal_point, = ax.plot(goal_position[0], goal_position[1], 'go', label='Goal')

# Path line
path_line, = ax.plot([], [], 'c-', linewidth=2, label='Path')

# Robot body
robot_body = Rectangle(
    (initial_position[0] - robot_size / 2, initial_position[1] - robot_size / 2),
    robot_size,
    robot_size,
    facecolor='red',
    edgecolor='red',
    label='Robot'
)
ax.add_patch(robot_body)

# Initial heading arrow
heading_arrow = ax.arrow(
    initial_position[0],
    initial_position[1],
    arrow_length,
    0.0,
    head_width=0.08,
    head_length=0.1,
    fc='purple',
    ec='purple'
)

ax.legend(loc='upper left')

# Store robot path
path_x = []
path_y = []

# =========================
# Helper function
# =========================
def update_robot_pose(x, y, angle_deg):
    global heading_arrow

    # Reset rectangle base position so its center is at (x, y)
    robot_body.set_x(x - robot_size / 2)
    robot_body.set_y(y - robot_size / 2)

    # Rotate around robot center
    t = transforms.Affine2D().rotate_deg_around(x, y, angle_deg) + ax.transData
    robot_body.set_transform(t)

    # Remove old arrow
    if heading_arrow is not None:
        heading_arrow.remove()

    # Draw new heading arrow
    angle_rad = math.radians(angle_deg)
    arrow_dx = arrow_length * math.cos(angle_rad)
    arrow_dy = arrow_length * math.sin(angle_rad)

    heading_arrow = ax.arrow(
        x,
        y,
        arrow_dx,
        arrow_dy,
        head_width=0.08,
        head_length=0.1,
        fc='purple',
        ec='purple'
    )

# =========================
# Init function
# =========================
def init():
    global heading_arrow, path_x, path_y

    path_x = [initial_position[0]]
    path_y = [initial_position[1]]

    update_robot_pose(initial_position[0], initial_position[1], 0.0)
    path_line.set_data(path_x, path_y)

    ax.set_title("Robot Navigation Simulation")
    return robot_body, path_line, heading_arrow

# =========================
# Animation update
# =========================
def update(frame):
    global path_x, path_y

    # Phase 1: Rotate in place from 0 deg to theta_deg
    if frame < rotate_frames:
        progress = frame / (rotate_frames - 1)
        current_angle = progress * theta_deg

        x = initial_position[0]
        y = initial_position[1]

        update_robot_pose(x, y, current_angle)

        path_line.set_data(path_x, path_y)
        ax.set_title(f"Rotating: {current_angle:.1f} deg")

    # Phase 2: Move from start to goal while keeping heading
    else:
        move_index = frame - rotate_frames
        progress = move_index / (move_frames - 1)

        x = initial_position[0] + progress * dx
        y = initial_position[1] + progress * dy

        update_robot_pose(x, y, theta_deg)

        # Update path
        if len(path_x) == 0 or (x != path_x[-1] or y != path_y[-1]):
            path_x.append(x)
            path_y.append(y)

        path_line.set_data(path_x, path_y)
        ax.set_title(f"Moving: x={x:.2f}, y={y:.2f}, distance={distance:.3f}")

    return robot_body, path_line, heading_arrow

# =========================
# Run animation
# =========================
ani = FuncAnimation(
    fig,
    update,
    frames=total_frames,
    init_func=init,
    interval=40,
    blit=False,
    repeat=False
)

# Hiển thị trong Jupyter
display(HTML(ani.to_jshtml()))

# =========================
# Save as GIF
# =========================
gif_filename = "robot_navigation.gif"
writer = PillowWriter(fps=25)
ani.save(gif_filename, writer=writer)

print(f"GIF saved successfully: {prefix/gif_filename}")
```
